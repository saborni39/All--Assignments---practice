```python
# 1.1 & 1.2 List_of_countries_by_continents
import requests
import csv
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame 

#Happiness report page
wikipage = requests.get("https://simple.wikipedia.org/wiki/List_of_countries_by_continents")

soup = BeautifulSoup(wikipage.content,"html.parser")
#print(soup)

data = []
data1 = []
data2 = []
data3 = [] 
data4 = []
data5 = []
value = ['Africa']
value1 = ["Asia"]
value2 = ["Europe"]
value3 = ["North America"]
value4 = ["South America"]
value5 = ["Australia/Oceania"]

for headline in soup('span', {'class' : 'mw-headline'}, id = 'Africa'):
    print(headline.text)
    links = headline.find_next('ol').find_all('a')
    for link in links:
        #if links[link] % 2 = 0
            all_content = link.text
            data.append(all_content) 
            data.append(value)
        
#all_content = link.text

    
#data frame df_happiness with all columns
df_countries = pd.DataFrame(data,columns=['Countries']) 

#to show all rows
pd.set_option('display.max_rows', None)
print(df_countries)

#####

for headline1 in soup('span', {'class' : 'mw-headline'}, id = 'Asia'):
    print(headline1.text)
    links1 = headline1.find_next('ol').find_all('a')
    for link1 in links1:
        all_content1 = link1.text
        data1.append(all_content1)
        data1.append(value1)

    
#data frame df_happiness with all columns
df_countries1 = pd.DataFrame(data1,columns=['Countries']) 

#to show all rows
pd.set_option('display.max_rows', None)
print(df_countries1)

####
for headline2 in soup('span', {'class' : 'mw-headline'}, id = 'Europe'):
    print(headline2.text)
    links2 = headline2.find_next('ol').find_all('a')
    for link2 in links2:
        all_content2 = link2.text
        data2.append(all_content2)
        data2.append(value2)
#all_content2 = link2.text

    
#data frame df_happiness with all columns
df_countries2 = pd.DataFrame(data2,columns=['Countries']) 

#to show all rows
pd.set_option('display.max_rows', None)
print(df_countries2)

######

for headline3 in soup('span', {'class' : 'mw-headline'}, id = 'North_America'):
    print(headline3.text)
    links3 = headline3.find_next('ol').find_all('a')
    for link3 in links3:
        all_content3 = link3.text
        data3.append(all_content3) 
        data3.append(value3)
#all_content3 = link3.text

    
#data frame df_happiness with all columns
df_countries3 = pd.DataFrame(data3,columns=['Countries']) 

#to show all rows
pd.set_option('display.max_rows', None)
print(df_countries3)

#####

for headline4 in soup('span', {'class' : 'mw-headline'}, id = 'South_America'):
    print(headline4.text)
    links4 = headline4.find_next('ol').find_all('a')
    for link4 in links4:
        all_content4 = link4.text
        data4.append(all_content4)
        data4.append(value4)
#all_content = link.text

    
#data frame df_happiness with all columns
df_countries4 = pd.DataFrame(data4,columns=['Countries']) 

#to show all rows
pd.set_option('display.max_rows', None)
print(df_countries4)

#####

for headline5 in soup('span', {'class' : 'mw-headline'}, id = 'Australia/Oceania'):
    print(headline5.text)
    links5 = headline5.find_next('ol').find_all('a')
    for link5 in links5:
        all_content5 = link5.text
        data5.append(all_content5) 
        data5.append(value5)
#all_content = link.text

    
#data frame df_happiness with all columns
df_countries5 = pd.DataFrame(data5,columns=['Countries']) 

#to show all rows
pd.set_option('display.max_rows', None)
print(df_countries5)

df = pd.DataFrame(columns=['Countries'])
final_frames = [df,df_countries,df_countries1,df_countries2,df_countries3,df_countries4,df_countries5]
#final_frames = df_countries['Countries'].astype(str) + '-' + df_countries1['Countries'].astype(str) + '-' + df_countries2['Countries'].astype(str)
result = pd.concat(final_frames)
print(result)
result.to_csv('Country-Continent Report.csv', header=False, index=False)
```
