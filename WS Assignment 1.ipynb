```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
r = np.random.uniform(0, 1000, 100)
print("Printing list of 100 random float numbers: ")
print (r)
#t=[0,30,45,60,90]
x=[i*(np.pi/180) for i in r]
SIN= np.sin(x)
COS = np.cos(x)
plt.scatter(x,SIN, marker="+",label="sin")
plt.scatter(x,COS,marker="*",label="cos")
plt.title("sin and cos values")
plt.xlabel("sine angle")
plt.ylabel("Cos angle")
plt.legend()
plt.show()
```
