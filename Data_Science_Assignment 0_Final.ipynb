```python
# 1.3 & 1.4 Happiness page
import requests
import csv
import numpy as np
import pandas as pd
#import urllib
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns

#Happiness report page
#address = "https://en.wikipedia.org/wiki/World_Happiness_Report"
#wikipage = urllib.request.urlopen(address)
wikipage = requests.get("https://en.wikipedia.org/wiki/World_Happiness_Report")

#soup = BeautifulSoup(wikipage,"html.parser")
soup = BeautifulSoup(wikipage.content,"html.parser")

#table explore
tables = soup.find_all("table")
h_table = soup.find('table', class_='wikitable sortable')
table_rows = h_table.find_all('tr')

data = []

#to find exact info
for row in table_rows:
    data.append([t.text.strip() for t in row.find_all('td')])
    
#data frame df_happiness with all columns
df_happiness = pd.DataFrame(data, columns=['Happiness_ranking', 'Country', 'score', 'GPD per capita', 'Social_support', 'Healthy_life_expectancy', 'Freedom_to_make_life_choices', 'Generosity', 'Perception_of_corruption']) 

#to show all rows
pd.set_option('display.max_rows', None)
print(df_happiness)

#to save it in a CSV
df_happiness.to_csv('Happiness Report.csv')


merged_frames = [df_happiness,result]
final_result = pd.concat(merged_frames)
print(final_result)

final_result.to_csv('Final Report(Happiness report & Countries report).csv')
```
